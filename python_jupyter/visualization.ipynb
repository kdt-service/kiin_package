```python
# %%
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
x = [10, 14, 19, 20, 25] 
plt.plot(x)
plt.show()
```

```python
x = np.arange(-4.5, 5, 0.5) 
y = x**2
plt.plot(x, y) 
plt.show()
```

```python
x=np.arange(-4.5,5,0.5)
y1=x**2 
y2=x+30 
y3=x**2+10
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
x = np.arange(-4.5, 5, 0.5) 
y1 = x**2
y2 = x+30 
y3 = x**2+10
plt.plot(x, y1, color="red", linestyle="dashed")
plt.plot(x, y2, "b:")
plt.plot(x, y3, color='#00FF00', linestyle="-.")
plt.show()
```

```python
x = np.linspace(0,10,100) 
y = x**2
plt.plot(x, y, '--r') 
plt.xlim(-20, 30) 
# plt.ylim(-20, 120)
# plt.axis([-20, 30, -20, 120]) 
plt.show()
```

```python
x = np.linspace(0,10,100) 
y1 = x**2
y2 = x**3 
# 
plt.plot(x, y1, 'r', label='line-1') 
plt.plot(x, y2, 'g', label='line-2')
# x,y 
plt.xlabel('x-label') 
plt.ylabel('y-label')
# plt.legend(loc=2) # legend  ( )
plt.legend(bbox_to_anchor=(1.04,1)) # legend ( ) 
plt.show()
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

penguins = sns.load_dataset("penguins")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```
