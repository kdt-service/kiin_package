### 라인그래프

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib

fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/NanumGothic.ttf' # For Windows
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
```

```python
x = [10, 14, 19, 20, 25]
plt.plot(x)
plt.show()
```

```python
x = np.arange(-4.5, 5.0, 0.5)
print(x)
print(type(x))

y = x**2
print(y)

# 리스트에서는 연산 불가
# [1, 2, 3, 4] ** 2
```

```python
plt.plot(x, y)
plt.show()
```

```python
y1 = x ** 2
y2 = x + 30
y3 = x ** 2 + 10
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

### 그래프 옵션

```python
plt.plot(x, y1, color="r", linestyle=":")
plt.plot(x, y2, '-b')
plt.plot(x, y3, color="#00ff00", linestyle="-.")

plt.show()
```

```python
x = np.linspace(0, 10, 100)

print(x)
print(type(x))
print(len(x))
```

```python
y = x**2

plt.plot(x, y, '--r')
# plt.xlim(-20, 20)
# plt.ylim(-10, 120)
plt.axis([-10, 20, -10, 120])
plt.show()
```

```python
x = np.linspace(0, 10, 100)

y1 = x**2
y2 = x**3
```

```python
plt.plot(x, y1, label="line-1")
plt.plot(x, y2, label="line-2")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.legend(bbox_to_anchor=[1.04, 1])
plt.show()
```

```python
plt.figure(figsize=[10, 6])

plt.subplot(221)
plt.plot([1, 2], color='r')

plt.subplot(222)
plt.plot([1, 2], color='g')

plt.subplot(223)
plt.plot([1, 2], color='b')

# plt.subplot(224)
# plt.plot([1, 2], color='y')

plt.show()
```

```python
plt.figure(figsize=(12, 5))
grid = plt.GridSpec(2, 3)

plt.subplot(grid[0, 0])
plt.plot([1, 2], color='r')

plt.subplot(grid[0, 1])
plt.plot([1, 2], color='g')

plt.subplot(grid[0, 2])
plt.plot([1, 2], color='b')

plt.subplot(grid[1, :2])
plt.plot([1, 2], color='y')

plt.subplot(grid[1, 2])
plt.plot([1, 2], color='m')

plt.show()
```

- 막대그래프

```python
labels = ['10s', '20s', '30s', '40s', '50s']

plt.bar(labels, [np.random.randint(1, 50) for _ in range(5)])

plt.title('bar-chart')
plt.xlabel('age')
plt.ylabel('count')

plt.show()
```
